```python
a=open("input1.txt","r")
num=a.readline()
inp=[]
for i in range(int(num)):
    b= a.readline().strip("\n")
    inp.append(b)
    
for i in range(len(inp)):
    if "@" in inp[i]:
        state=0
        j=0
        if inp[i][j].isalpha():
            state=1
            j+=1
        else:
            state="dead"
            print("Invalid Email")
            continue
        server=["gmail","outlook","yahoo","bracu"]
        domain=['com',"net","edu",'bd','org']
        strr=""
        count=0
        while True:
            if inp[i][j]=="@":
                state+=1
                j+=1
                break
            if inp[i][j].isalpha() or inp[i][j].isnumeric() or inp[i][j]==".":
                state=2
                j+=1
                continue
            else:
                count=1
                break 
           
        if count==1:
            print("Invalid Email")
            continue
            
        c,d=inp[i][j:].split(".")

        if c in server:
            state+=1
        if d in domain:
            state+=1
        if state==5:
            print("Email,",i+1)
        else:
            print("Invalid Email")
            
    elif "http://" in inp[i][:7] or "https://" in inp [i][:8] or "www" in inp[i][:3]:
        
        state_web=0
        j=0
        protocal=""
        while True:
            if inp[i][j]==".":
                j+=1
                break
            protocal+=inp[i][j]
            j+=1
        
        protocl=["http://","https://","www","https://www","https://web"]
        
        if protocal in protocl:
            state_web+=1
        else:
            state_web="dead"
            print("Invalid Web")
            continue
                
        while True:
            if inp[i][j]==".":
                state_web+=1
                j+=1
                break
            if "a"<= inp[i][j] <="z":
                state_web=2
                j+=1
            else:
                state_web="dead"
                print("Invalid Web")
                break
        if state_web=="dead":
            continue
        domain=['com',"net","edu",'org']
        
        if inp[i][j:] in domain:
            state_web+=1
            
        if state_web==4:
            print("Web,",i+1)
        else:
            print("Invalid web")
    else:
        print("Invalid mail or web")
            
```
